```python
import pandas as pd
import glob
import os
import shutil
```

```python
file = pd.read_csv("PhotoImageInformation.csv")
print(file.shape)
file = file.dropna(subset=["DPI Information"])
print(file.shape)
file[["DPI Info 1","DPI Info 2"]] = file["DPI Information"].str.split(",",expand=True)
file["DPI Info 1"] = file["DPI Info 1"].str.replace("(","")
file["DPI Info 2"] = file["DPI Info 2"].str.replace(")","")
file["Folderpath"] = "photo/" + file["Filename"].str.split("/").str[1] + "/" + file["Filename"].str.split("/").str[-1].str.split(".").str[0]
file["DPI Info 1"] = file["DPI Info 1"].str.replace("72.0","72")
file["DPI Info 2"] = file["DPI Info 2"].str.replace("72.0","72")
file["DPI Info 1"] = file["DPI Info 1"].astype(int)
file["DPI Info 2"] = file["DPI Info 2"].astype(int)
file.head()
```

```python
dst_folder = "filtered_dataset"
list_of_files = []
for index,rows in file.iterrows():
    if rows["DPI Info 1"] > 150:
        list_of_files.append(rows["Folderpath"])
print(list_of_files)
```

```python
for files in list_of_files:
    file_name = os.path.basename(files)
    if file_name + ".jpg":
        try:
            shutil.move(files+".jpg",dst_folder)
            print(files + "found of jpg format")
        except shutil.Error:
            pass
    if file_name + ".xml":
        try:
            shutil.move(files+".xml",dst_folder)
            print(files + "found of xml format")
        except shutil.Error:
            pass
    else:
        print(files + "not found")
```
