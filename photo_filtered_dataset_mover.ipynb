```python
import pandas as pd
import glob
import os
import shutil
```

```python
file = pd.read_csv("D:/Thesis/PlantCLEF2020TrainingData2/Ukiyoe_information.csv")
print(file.shape)
file.head()
```

```python
dir_path = "D:/Thesis/realism_data/Ukiyo-e_filtered_dataset"
if not os.path.exists(dir_path):
      os.makedirs(dir_path)
      print("Directory created successfully!")
else:
      print("Directory already exists!")
list_of_files = []
for index,rows in file.iterrows():
    if rows["Width"] >= 384 and rows["Height"] >= 384:
        list_of_files.append(rows["Filename"])
print(len(list_of_files))
```

```python
list_of_files
```

```python
for files in list_of_files:
    file_name = os.path.basename(files)
    dest_path = os.path.join(dir_path,file_name)
    if os.path.exists(files):
        try:
            shutil.copy(files, dir_path)
            print(files + " found of jpg format")
        except shutil.Error:
            print(files + " already exists in destination.")
    else:
        print(files + " not found")
```
